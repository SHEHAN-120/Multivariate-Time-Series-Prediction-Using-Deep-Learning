```python
def evaluate_model(name, y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"{name} Performance")
    print(f"RMSE: {rmse:.4f}")
    print(f"MAE:  {mae:.4f}")
    print(f"R2:   {r2:.4f}")
    return y_pred
```

```python
def plot_results(y_true, y_pred, title):
    plt.figure(figsize=(14, 6))
    
    plt.subplot(1, 2, 1)
    plt.plot(y_true[:100], label='Actual', color='blue')
    plt.plot(y_pred[:100], label='Predicted', color='red', linestyle='--')
    plt.title(f'{title}: Actual vs Predicted (Begining to 100 steps)')
    plt.legend()
   
    residuals = y_true - y_pred.flatten()
    plt.subplot(1, 2, 2)
    sns.histplot(residuals, kde=True)
    plt.title(f'{title}: Residual Distribution')
    plt.xlabel("Error")
    plt.tight_layout()
    plt.show()
```

```python
def main():
    
    BASE_DIR = os.getcwd()
    
    
    DATA_PATH = os.path.join(BASE_DIR, '../data/raw/energy_data_set.csv')
    MODEL_DIR = os.path.join(BASE_DIR, 'models')

    
    if not os.path.exists(DATA_PATH):
        print(f"ERROR: Data file not found at {DATA_PATH}")
        print("Current working directory is:", BASE_DIR)
        return

    if not os.path.exists(MODEL_DIR):
        os.makedirs(MODEL_DIR)
        print(f"Created directory: {MODEL_DIR}")

    print(f"Loading data from {DATA_PATH}...")
    df = data_preprocessing.load_data(DATA_PATH)

    print("Removing Outliers")
    df = data_preprocessing.remove_outliers_iqr(df)
    
    df = ab.feature_engineering_pipeline(df)

    train_scaled, test_scaled, scaler = data_preprocessing.split_and_scale(df)
    target_idx = list(train_scaled.columns).index('Appliances')
    
    WINDOW = 24

    X_train, y_train = ab.create_sequences(train_scaled, target_idx, WINDOW)
    X_test, y_test = ab.create_sequences(test_scaled, target_idx, WINDOW)
    
    print("Training Baseline: Linear Regression.")
    lr_model = model.train_baseline_lr(X_train, y_train)

    lr_preds = lr_model.predict(X_test.reshape(X_test.shape[0], -1))
    evaluate_model("Linear Regression", y_test, lr_preds)
    
    print("Training Baseline: Random Forest.")
    rf_model = model.train_baseline_rf(X_train, y_train)
    rf_preds = rf_model.predict(X_test.reshape(X_test.shape[0], -1))
    evaluate_model("Random Forest", y_test, rf_preds)
    
    print("Training Deep Learning: LSTM.")
    lstm_model = model.build_lstm_model((X_train.shape[1], X_train.shape[2]))
    
    checkpoint_path = os.path.join(MODEL_DIR, 'best_model.h5')
    checkpoint = ModelCheckpoint(checkpoint_path, monitor='val_loss', save_best_only=True, mode='min', verbose=1)
    early_stop = EarlyStopping(patience=3, restore_best_weights=True)

    lstm_model.fit(
        X_train, y_train,
        validation_split=0.2,
        epochs=30, 
        batch_size=64, 
        callbacks=[early_stop, checkpoint], 
        verbose=1 ) 
    
    final_model_path = os.path.join(MODEL_DIR, 'trained_model.h5') 
    lstm_model.save(final_model_path) 
    print(f"Final model successfully saved to: {final_model_path}") 

    SCALER_PATH = os.path.join(MODEL_DIR, 'scaler.pkl')
    joblib.dump(scaler, SCALER_PATH)
    print(f"Scaler successfully saved to: {SCALER_PATH}")
    
    lstm_preds = lstm_model.predict(X_test) 
    evaluate_model("LSTM", y_test, lstm_preds)
    plot_results(y_test, lstm_preds, "LSTM_Model") 

if __name__ == "__main__": 
    main()
```
