```python
import numpy as np
import pandas as pd
```

```python
def feature_engineering_pipeline(df):
    df = df.copy()
    
    
    df['month'] = df['date'].dt.month
    df['weekday'] = df['date'].dt.weekday
    df['hour'] = df['date'].dt.hour
    df['week'] = df['date'].dt.isocalendar().week
    df['Appliances_rolling_mean'] = df['Appliances'].shift(1).rolling(window=6).mean()

    df['lag_1'] = df['Appliances'].shift(1)
    df['lag_3'] = df['Appliances'].shift(3)
    df['lag_6'] = df['Appliances'].shift(6)

    df['T_out_RH_out'] = df['T_out'] * df['RH_out']
    df['T1_RH1'] = df['T1'] * df['RH_1']
    
    df = df.fillna(method='bfill').fillna(method='ffill')
    return df
```

```python
def create_sequences(input_data, target_col_idx, window_size=24):
    X, y = [], []
    data_array = input_data.values
    for i in range(len(data_array) - window_size):
        X.append(data_array[i : i + window_size])
        y.append(data_array[i + window_size][target_col_idx])
    return np.array(X), np.array(y)
```
